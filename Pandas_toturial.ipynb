# DataFrame Creation and Manipulation:
pd.DataFrame(): Create a DataFrame from various data sources.

df.head(), df.tail(): Display the first/last rows of a DataFrame.

df.info(), df.describe(): Display information and summary statistics.

df.shape, df.size: Get dimensions and total number of elements.

df.columns, df.index: Access column and index labels.

df.drop(): Remove rows or columns from a DataFrame.

df.rename(): Rename columns or index labels.

df.sort_values(), df.sort_index(): Sort DataFrame by values or index.

df.merge(), df.join(): Merge and join DataFrames.

df.pivot(), df.pivot_table(): Reshape data using pivoting.

df.melt(): Unpivot/melt DataFrame.

df.apply(), df.applymap(): Apply functions to elements or DataFrame.

```python
import pandas as pd

# DataFrame Creation and Manipulation
# Create a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Display the first few rows of the DataFrame
print("\nFirst 2 rows:\n", df.head(2))

# Display the last few rows of the DataFrame
print("\nLast 2 rows:\n", df.tail(2))

# Display summary information about the DataFrame
df.info()

# Display summary statistics of numeric columns
print("\nSummary statistics:\n", df.describe())

# Get the dimensions (rows, columns) of the DataFrame
print("\nShape of DataFrame:", df.shape)

# Get the total number of elements in the DataFrame
print("Total number of elements:", df.size)

# Access column labels
print("\nColumn labels:", df.columns)

# Access index labels
print("Index labels:", df.index)

# Remove a column from the DataFrame
df_dropped = df.drop(columns=['Age'])
print("\nDataFrame after dropping 'Age' column:\n", df_dropped)

# Rename columns
df_renamed = df.rename(columns={'Name': 'Full Name', 'Age': 'Years'})
print("\nDataFrame with renamed columns:\n", df_renamed)

# Sort DataFrame by values in the 'Age' column
df_sorted = df.sort_values(by='Age', ascending=False)
print("\nDataFrame sorted by 'Age':\n", df_sorted)

# Merge DataFrames based on a common column
data2 = {'Name': ['David', 'Eve'],
         'Country': ['USA', 'Australia']}
df2 = pd.DataFrame(data2)
merged_df = df.merge(df2, on='Name')
print("\nMerged DataFrame:\n", merged_df)

# Pivot the DataFrame
pivot_df = df.pivot(index='Name', columns='Age')
print("\nPivoted DataFrame:\n", pivot_df)

# Apply a function element-wise to DataFrame
def add_suffix(name):
    return name + "_New"
df_with_suffix = df.applymap(add_suffix)
print("\nDataFrame with added suffix:\n", df_with_suffix)
```


# Data Selection and Indexing:
df[column_name]: Access a column as a Series.

df.loc[], df.iloc[]: Access data by labels or integer positions.

df.at[], df.iat[]: Access single values using labels or positions.

df.filter(): Select columns based on labels, strings, or regex.

df.query(): SQL-like queries for data selection.

df.isin(), df.isnull(), df.notnull(): Filter data based on conditions.

df.sample(): Randomly sample rows from DataFrame.

df.set_index(), df.reset_index(): Set/reset DataFrame index.

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'Country': ['USA', 'Canada', 'UK']}
df = pd.DataFrame(data)

# Access a column as a Series
name_column = df['Name']
print("Name Column:\n", name_column)

# Access data by labels using loc[]
alice_info = df.loc[0]
print("\nAlice's Info using loc[]:\n", alice_info)

# Access data by integer positions using iloc[]
bob_info = df.iloc[1]
print("\nBob's Info using iloc[]:\n", bob_info)

# Access a single value using at[]
age_of_charlie = df.at[2, 'Age']
print("\nCharlie's Age using at[]:", age_of_charlie)

# Access a single value using iat[]
age_of_bob_iat = df.iat[1, 1]
print("Bob's Age using iat[]:", age_of_bob_iat)

# Select columns based on labels using filter()
selected_columns = df.filter(items=['Name', 'Age'])
print("\nSelected Columns using filter():\n", selected_columns)

# Select columns based on regex using filter()
name_columns = df.filter(like='Name')
print("\nColumns with 'Name' using filter(like='Name'):\n", name_columns)

# SQL-like queries for data selection using query()
uk_citizens = df.query("Country == 'UK'")
print("\nUK Citizens using query():\n", uk_citizens)

# Filter data based on conditions using isin()
filtered_data = df[df['Age'].isin([25, 22])]
print("\nFiltered data using isin():\n", filtered_data)

# Check for missing values using isnull() and notnull()
null_age = df[df['Age'].isnull()]
not_null_country = df[df['Country'].notnull()]
print("\nRows with null Age:\n", null_age)
print("\nRows with not null Country:\n", not_null_country)

# Randomly sample rows from DataFrame using sample()
sampled_rows = df.sample(n=2)
print("\nSampled Rows:\n", sampled_rows)

# Set 'Name' column as the index using set_index()
df_indexed = df.set_index('Name')
print("\nDataFrame with 'Name' as Index:\n", df_indexed)

# Reset index using reset_index()
df_reset = df_indexed.reset_index()
print("\nDataFrame with Reset Index:\n", df_reset)

```


# Data Cleaning and Transformation:

df.fillna(): Fill missing values with specified values or methods.

df.replace(): Replace values with other values.

df.duplicated(), df.drop_duplicates(): Identify and remove duplicates.

df.astype(), df.convert_dtypes(): Convert data types of columns.

df.cut(), df.qcut(): Bin values into discrete intervals.

df.dropna(): Remove rows or columns with missing values.

df.dropna(thresh=n): Remove rows with fewer than n non-NA values.

df.dropna(subset=[col1, col2]): Remove rows with NA values in specific columns.

```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, np.nan, 22, 28, 35],
        'Country': ['USA', 'Canada', 'UK', np.nan, 'Australia']}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Fill missing values with specified values or methods using fillna()
df_filled = df.fillna({'Age': 30, 'Country': 'Unknown'})
print("\nDataFrame after filling missing values:\n", df_filled)

# Replace values using replace()
df_replaced = df_filled.replace('Canada', 'CA')
print("\nDataFrame after value replacement:\n", df_replaced)

# Identify and remove duplicates using duplicated() and drop_duplicates()
duplicated_rows = df_replaced.duplicated(subset=['Name'])
df_unique = df_replaced[~duplicated_rows]
print("\nDataFrame with duplicates removed:\n", df_unique)

# Convert data types of columns using astype()
df_converted = df_unique.astype({'Age': 'int'})
print("\nDataFrame with converted data types:\n", df_converted.dtypes)

# Bin values into discrete intervals using cut()
age_bins = [0, 20, 30, 40]
age_labels = ['0-20', '21-30', '31-40']
df_binned = df_converted.copy()
df_binned['Age Group'] = pd.cut(df_binned['Age'], bins=age_bins, labels=age_labels)
print("\nDataFrame with age bins:\n", df_binned)

# Remove rows with missing values using dropna()
df_no_missing = df_binned.dropna()
print("\nDataFrame after dropping rows with missing values:\n", df_no_missing)

# Remove rows with fewer than n non-NA values using dropna(thresh=n)
df_min_non_na = df_binned.dropna(thresh=2)
print("\nDataFrame after dropping rows with less than 2 non-NA values:\n", df_min_non_na)

# Remove rows with NA values in specific columns using dropna(subset=[col1, col2])
df_no_na_age = df_binned.dropna(subset=['Age'])
print("\nDataFrame after dropping rows with NA values in 'Age' column:\n", df_no_na_age)

```



# Data Aggregation and Calculation:

df.mean(), df.median(), df.sum(), df.min(), df.max(): Aggregate functions.

df.count(), df.nunique(): Count non-NA/null values and unique values.

df.groupby(): Group data by one or more columns.

df.agg(), df.transform(): Apply multiple aggregation or transformation functions.

df.rank(), df.cumsum(), df.diff(): Compute ranks, cumulative sums, differences.

df.corr(), df.cov(): Compute correlation and covariance matrices.


```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 22, 28, 35],
        'Salary': [50000, 60000, 55000, 62000, 58000],
        'Country': ['USA', 'Canada', 'UK', 'USA', 'Australia']}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Aggregate functions
mean_age = df['Age'].mean()
median_salary = df['Salary'].median()
total_salary = df['Salary'].sum()
min_age = df['Age'].min()
max_age = df['Age'].max()
print("\nMean Age:", mean_age)
print("Median Salary:", median_salary)
print("Total Salary:", total_salary)
print("Minimum Age:", min_age)
print("Maximum Age:", max_age)

# Count non-NA/null values and unique values
non_na_count = df.count()
unique_countries = df['Country'].nunique()
print("\nNon-NA Count:\n", non_na_count)
print("Number of Unique Countries:", unique_countries)

# Group data by 'Country' and calculate mean salary
grouped_by_country = df.groupby('Country')['Salary'].mean()
print("\nMean Salary by Country:\n", grouped_by_country)

# Apply multiple aggregation functions using agg()
aggregation_functions = {
    'Age': ['mean', 'max'],
    'Salary': 'sum'
}
aggregated_data = df.groupby('Country').agg(aggregation_functions)
print("\nAggregated Data:\n", aggregated_data)

# Compute rank of 'Salary' column
df['Salary Rank'] = df['Salary'].rank(ascending=False)
print("\nDataFrame with Salary Ranks:\n", df)

# Compute cumulative sum of 'Salary' column
df['Salary Cumulative Sum'] = df['Salary'].cumsum()
print("\nDataFrame with Cumulative Salary Sums:\n", df)

# Compute differences between consecutive 'Age' values
df['Age Difference'] = df['Age'].diff()
print("\nDataFrame with Age Differences:\n", df)

# Compute correlation and covariance matrices
correlation_matrix = df[['Age', 'Salary']].corr()
covariance_matrix = df[['Age', 'Salary']].cov()
print("\nCorrelation Matrix:\n", correlation_matrix)
print("\nCovariance Matrix:\n", covariance_matrix)
```


# Data Visualization:

df.plot(): Create basic plots using Matplotlib integration.

df.plot.scatter_matrix(): Pairwise scatter plot matrix.

df.plot.bar(), df.plot.line(), df.plot.box(): Different types of plots.

df.plot.hist(), df.plot.kde(), df.plot.area(): Histograms, kernel density estimates, area plots.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 22, 28, 35],
        'Salary': [50000, 60000, 55000, 62000, 58000],
        'Country': ['USA', 'Canada', 'UK', 'USA', 'Australia']}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Create basic plots using Matplotlib integration
df.plot(x='Name', y='Age', kind='bar')
plt.title("Age of Individuals")
plt.xlabel("Name")
plt.ylabel("Age")
plt.show()

# Create a pairwise scatter plot matrix
from pandas.plotting import scatter_matrix
scatter_matrix(df[['Age', 'Salary']], figsize=(8, 8))
plt.show()

# Create different types of plots
df.plot(x='Name', y='Salary', kind='line')
plt.title("Salary of Individuals")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.show()

df.plot(x='Name', y='Age', kind='box')
plt.title("Age Distribution")
plt.show()

# Create histograms, kernel density estimates, and area plots
df.plot(y='Age', kind='hist', bins=10, alpha=0.7)
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

df.plot(y='Age', kind='kde')
plt.title("Kernel Density Estimation")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()

df.plot(y='Salary', kind='area', alpha=0.4)
plt.title("Salary Area Plot")
plt.xlabel("Individuals")
plt.ylabel("Salary")
plt.show()

    
```



# Time Series and Datetime Handling:

pd.to_datetime(): Convert strings to datetime objects.

df.resample(): Resample time series data.

df.shift(), df.diff(): Compute time-lagged differences.

df.rolling(): Calculate rolling window statistics.

```python
import pandas as pd

# Sample time series DataFrame
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Value': [10, 15, 20, 25, 30]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print("Original Time Series DataFrame:\n", df)

# Convert strings to datetime objects using pd.to_datetime()
date_strings = ['2023-01-06', '2023-01-07', '2023-01-08']
datetime_objects = pd.to_datetime(date_strings)
print("\nConverted Datetime Objects:\n", datetime_objects)

# Resample time series data using df.resample()
resampled_df = df.resample('D').sum()  # Resample to daily frequency
print("\nResampled Time Series DataFrame (Daily):\n", resampled_df)

# Compute time-lagged differences using df.shift() and df.diff()
df_shifted = df.shift(periods=1)
df_diff = df.diff(periods=1)
print("\nShifted Time Series DataFrame:\n", df_shifted)
print("\nDifferences Time Series DataFrame:\n", df_diff)

# Calculate rolling window statistics using df.rolling()
rolling_mean = df['Value'].rolling(window=3).mean()
rolling_sum = df['Value'].rolling(window=2).sum()
print("\nRolling Mean Time Series:\n", rolling_mean)
print("\nRolling Sum Time Series:\n", rolling_sum)
```
